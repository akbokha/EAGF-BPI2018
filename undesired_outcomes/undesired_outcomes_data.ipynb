### Case lifecycle

So there are 43809 cases present in the event log. Let's see how many of these cases are actually useful to use for training/testing/cross-validation.

```python
elog.loc[elog['Activity'] == 'Payment application-Application-decide'].shape
```

```python
len(set(elog.loc[elog['Activity'] == 'Payment application-Application-decide']['Case ID']))
```

We see that there are only 43809 - 43793 = 16 cases that do not have the the 'Payment application-Application-decide' activity in their lifecycle (as far as can be deduced from the event logs)

Let's drop these 16 cases since they are incomplete in the context of this business question and are not that significant when one considers the number of cases at hand.

```python
to_be_dropped_cases = set(elog['Case ID']).difference(set(set(elog.loc[elog['Activity'] == 'Payment application-Application-decide']['Case ID'])))
```

```python
elog = elog[~elog['Case ID'].isin(to_be_dropped_cases)]
```

```python
len(set(elog['Case ID'])) # for confirmation
```

But since we want to be able to detect the undesired outcomes before the 'Payment application-Application-decide' the choice is made to consider the event that proceeds this activity in almost all the cases: 'Payment application-Application-finish editing'.

```python
len(set(elog.loc[elog['Activity'] == 'Payment application-Application-finish editing']['Case ID']))
```

```python
elog_uo = elog[elog['Activity'] == 'Payment application-Application-finish editing']
```

### Undesired outcome #1

```python
set(elog['Activity'])
```

```python
def is_undesired_outcome1(x):
    case_id = x['Case ID']
    case_subset = elog[elog['Case ID'] == case_id]
    case_events = set(case_subset['Activity'])
    if 'Payment application-Application-begin payment' in case_events: # there has been a begin payment activity
        if 'Payment application-Application-abort payment' not in case_events: # no abort payment activity in the data
            return True
        else: # check whether it was before the end of the year
            earliest_date = min(list(case_subset['timestamp_year']))
            date_abort_payment = list(case_subset[case_subset['Activity'] == 
                                             'Payment application-Application-abort payment']['timestamp_year'])[0]
            if date_abort_payment > earliest_date:
                return True
    return False
```

```python
elog_uo['undesired_outcome1'] = elog_uo.apply(is_undesired_outcome1, axis=1)
```

```python
print("shape ", elog_uo.shape[0])
print("true " , elog_uo[elog_uo['undesired_outcome1'] == True].shape[0])
print("false " , elog_uo[elog_uo['undesired_outcome1'] == False].shape[0])
```

### Undesired outcome #2

```python
def is_undesired_outcome2(x):
    case_id = x['Case ID']
    case_subset = elog[elog['Case ID'] == case_id]
    case_subprocesses = set(case_subset['subprocess'])
    if 'Change' in case_subprocesses or 'Objection' in case_subprocesses:
        return True
    else:
        return False
```

```python
elog_uo['undesired_outcome2'] = elog_uo.apply(is_undesired_outcome2, axis=1)
```

```python
elog_uo['undesired_outcome'] = np.where(elog_uo['undesired_outcome1'] | elog_uo['undesired_outcome2'], True, False)
```

```python
print("shape ", elog_uo.shape[0])
print("true " , elog_uo[elog_uo['undesired_outcome1'] == True].shape[0])
print("false " , elog_uo[elog_uo['undesired_outcome1'] == False].shape[0])
```

```python
elog_uo.to_csv(dest_dir.__add__('elog_uo').__add__(extension), index=False)
```
