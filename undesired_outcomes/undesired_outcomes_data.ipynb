### Case lifecycle

```python
len(set(elog['Case ID']))
```

So there are 43809 cases present in the event log. Let's see how many of these cases are actually useful to use for training/testing/cross-validation.

```python
elog.loc[elog['Activity'] == 'Payment application-Application-decide'].shape
```

```python
len(set(elog.loc[elog['Activity'] == 'Payment application-Application-decide']['Case ID']))
```

We see that there are only 43809 - 43793 = 16 cases that do not have the the 'Payment application-Application-decide' activity in their lifecycle (as far as can be deduced from the event logs)

Let's drop these 16 cases since they are incomplete in this context and are not that significant when you consider the number of cases at hand.

```python
to_be_dropped_cases = set(elog['Case ID']).difference(set(set(elog.loc[elog['Activity'] == 'Payment application-Application-decide']['Case ID'])))
```

```python
elog = elog[~elog['Case ID'].isin(to_be_dropped_cases)]
```

But since we want to be able to detect the undesired outcomes before the 'Payment application-Application-decide' the choice is made to consider the event that proceeds this activity in almost all the cases: 'Payment application-Application-finish editing'.

```python
len(set(elog.loc[elog['Activity'] == 'Payment application-Application-finish editing']['Case ID']))
```
