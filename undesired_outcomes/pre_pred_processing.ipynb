### amount applied processing

```python
def return_num_changes_applied_amount(x):
    result = 0
    if not (pd.isnull(x['(case) amount_applied1']) or x['(case) amount_applied0'] == x['(case) amount_applied1']):
        result += 1
    if not (pd.isnull(x['(case) amount_applied2']) or x['(case) amount_applied1'] == x['(case) amount_applied2']):
        result += 1
    if not (pd.isnull(x['(case) amount_applied3']) or x['(case) amount_applied2'] == x['(case) amount_applied3']):
        result += 1
    return result
```

```python
def return_final_applied_amount(x):
    if not pd.isnull(x['(case) amount_applied3']):
        return x['(case) amount_applied3']
    elif not pd.isnull(x['(case) amount_applied2']):
        return x['(case) amount_applied2']
    elif not pd.isnull(x['(case) amount_applied1']):
        return x['(case) amount_applied1']
    else:
        return x['(case) amount_applied0']
```

```python
def return_diff_applied_amount(x):
    if not pd.isnull(x['(case) amount_applied3']):
        return x['(case) amount_applied3'] - x['(case) amount_applied0']
    elif not pd.isnull(x['(case) amount_applied2']):
        return x['(case) amount_applied2'] -x['(case) amount_applied0']
    elif not pd.isnull(x['(case) amount_applied1']):
        return x['(case) amount_applied1'] - x['(case) amount_applied0']
    else:
        return 0
```

```python
# elog.loc[elog['(case) amount_applied1'] > 0]
```

```python
elog['num changes applied amount'] = elog.apply(return_num_changes_applied_amount, axis=1)
```

```python
# random control checks
elog.loc[elog['num changes applied amount'] > 0].sample(3)
elog.loc[elog['num changes applied amount'] > 1].sample(3)
elog.loc[elog['num changes applied amount'] > 2].sample(3)
```

```python
elog['final applied amount'] = elog.apply(return_final_applied_amount, axis=1)
```

```python
elog['diff applied amount'] = elog.apply(return_diff_applied_amount, axis=1)
```

```python
elog['applied amount lowered'] = np.where(elog['diff applied amount'] < 0, True, False)
```

```python
elog['applied amount changed'] = np.where(~pd.isnull(elog['(case) amount_applied1']), True, False)
```

### actual payment processing

```python
def return_num_changes_actual_amount(x):
    result = 0
    if not (pd.isnull(x['(case) payment_actual0']) or x['(case) payment_actual0'] == x['(case) payment_actual1']):
        result += 1
    if not (pd.isnull(x['(case) payment_actual2']) or x['(case) payment_actual1'] == x['(case) payment_actual2']):
        result += 1
    if not (pd.isnull(x['(case) payment_actual3']) or x['(case) payment_actual2'] == x['(case) payment_actual3']):
        result += 1
    return result
```

```python
def return_final_actual_amount(x):
    if not pd.isnull(x['(case) payment_actual3']):
        return x['(case) payment_actual3']
    elif not pd.isnull(x['(case) payment_actual2']):
        return x['(case) payment_actual2']
    elif not pd.isnull(x['(case) payment_actual1']):
        return x['(case) payment_actual1']
    else:
        return x['(case) payment_actual0']
```

```python
def return_diff_actual_amount(x):
    if not pd.isnull(x['(case) payment_actual3']):
        return x['(case) payment_actual3'] - x['(case) payment_actual0']
    elif not pd.isnull(x['(case) payment_actual2']):
        return x['(case) payment_actual2'] -x['(case) payment_actual0']
    elif not pd.isnull(x['(case) payment_actual1']):
        return x['(case) payment_actual1'] - x['(case) payment_actual0']
    else:
        return 0
```

```python
elog['num changes actual amount'] = elog.apply(return_num_changes_actual_amount, axis=1)
```

```python
elog['final actual amount'] = elog.apply(return_final_actual_amount, axis=1)
```

```python
elog['diff actual amount'] = elog.apply(return_diff_actual_amount, axis=1)
```

```python
elog['actual amount lowered'] = np.where(elog['diff actual amount'] < 0, True, False)
elog['actual amount changed'] = np.where(~pd.isnull(elog['(case) payment_actual1']), True, False)
elog['actual amount pay'] = np.where(elog['final actual amount'] < 0, True, False)
```

### penalty amount processing

```python
def return_num_changes_penalty_amount(x):
    result = 0
    if not (pd.isnull(x['(case) penalty_amount0']) or x['(case) penalty_amount0'] == x['(case) penalty_amount1']):
        result += 1
    if not (pd.isnull(x['(case) penalty_amount2']) or x['(case) penalty_amount1'] == x['(case) penalty_amount2']):
        result += 1
    if not (pd.isnull(x['(case) penalty_amount3']) or x['(case) penalty_amount2'] == x['(case) penalty_amount3']):
        result += 1
    return result

def return_final_penalty_amount(x):
    if not pd.isnull(x['(case) penalty_amount3']):
        return x['(case) penalty_amount3']
    elif not pd.isnull(x['(case) penalty_amount2']):
        return x['(case) penalty_amount2']
    elif not pd.isnull(x['(case) penalty_amount1']):
        return x['(case) penalty_amount1']
    else:
        return x['(case) penalty_amount0']

def return_diff_penalty_amount(x):
    if not pd.isnull(x['(case) penalty_amount3']):
        return x['(case) penalty_amount3'] - x['(case) penalty_amount0']
    elif not pd.isnull(x['(case) penalty_amount2']):
        return x['(case) penalty_amount2'] -x['(case) penalty_amount0']
    elif not pd.isnull(x['(case) penalty_amount1']):
        return x['(case) penalty_amount1'] - x['(case) penalty_amount0']
    else:
        return 0
```

```python
elog['num changes penalty amount'] = elog.apply(return_num_changes_penalty_amount, axis=1)
elog['final penalty amount'] = elog.apply(return_final_penalty_amount, axis=1)
elog['diff penalty amount'] = elog.apply(return_diff_penalty_amount, axis=1)
```

```python
elog['penalty amount lowered'] = np.where(elog['diff penalty amount'] < 0, True, False)
elog['penalty amount changed'] = np.where(~pd.isnull(elog['(case) penalty_amount1']), True, False)
elog['penalty amount corrected'] = np.where(elog['final penalty amount'] >= 0, True, False)
```
