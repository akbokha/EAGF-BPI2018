```python
import pandas as pd 
import numpy as np
```

```python
import bitstring
```

```python
source_dir = '../data/disco_to_csv/'
dest_dir = '../data/proc/'
extension = '.csv'
```

```python
# Complete event log
elog = pd.read_csv(source_dir.__add__('BPI Challenge 2018.csv'))
```

```python
# Subprocesses/documents
control_summary = pd.read_csv(source_dir.__add__('Control summary.csv'))
dep_control = pd.read_csv(source_dir.__add__('Department control parcels.csv'))
entitle_app = pd.read_csv(source_dir.__add__('Entitlement application.csv'))
geo_parcel = pd.read_csv(source_dir.__add__('Geo parcel document.csv'))
inspec = pd.read_csv(source_dir.__add__('Inspection.csv'))
parcel_doc = pd.read_csv(source_dir.__add__('Parcel document.csv'))
pay_app = pd.read_csv(source_dir.__add__('Payment application.csv'))
ref_align = pd.read_csv(source_dir.__add__('Reference alignment.csv'))
```

```python
all_subprocess = [control_summary, dep_control, entitle_app, geo_parcel, inspec, parcel_doc, pay_app, ref_align]
all_proc = all_subprocess + [elog]
```

```python
for sp in all_proc:
    sp.loc[sp['Resource'] == '0;n/a', 'Resource'] = np.NaN
```

```python
for sp in all_subprocess:
    sp['Case ID'] = sp['Case ID'].apply(lambda x: x if np.isnan(x) else bitstring.BitArray('int:64=%d' % x).hex)
```

```python
for sp in all_proc:
    sp['docid'] = sp['docid'].apply(lambda x: x if np.isnan(x) else bitstring.BitArray('int:64=%d' % x).hex)
    sp['eventid'] = sp['eventid'].apply(lambda x: x if np.isnan(x) else bitstring.BitArray('int:64=%d' % x).hex)
```
